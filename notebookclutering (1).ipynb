# **Code Python cho K-means Clustering:**

```python
# Import các thư viện cần thiết
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Tạo dữ liệu mẫu (bạn có thể thay đổi số lượng mẫu và cụm)
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=0)

# Trực quan hóa dữ liệu ban đầu
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Dữ liệu ban đầu")
plt.show()

# Khởi tạo và huấn luyện mô hình K-means
kmeans = KMeans(n_clusters=4, random_state=0)  # Đặt số cụm là 4
kmeans.fit(X)

# Dự đoán nhãn cụm cho từng điểm dữ liệu
labels = kmeans.predict(X)

# Tọa độ các tâm cụm
centroids = kmeans.cluster_centers_

# Trực quan hóa kết quả phân cụm
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', marker='X', label='Centroids')
plt.title("K-means Clustering")
plt.legend()
plt.show()
```

# **Giải thích từng phần:**
1.   **Import thư viện:**

*   `numpy`: Để xử lý các phép toán với mảng.

*   `matplotlib`: Để vẽ biểu đồ.

*   `sklearn.datasets`: Để tạo dữ liệu mẫu.

*   `sklearn.cluster`: Chứa hàm KMeans để phân cụm.


2.   **Tạo dữ liệu mẫu:**

*   Hàm `make_blobs` tạo ra dữ liệu ngẫu nhiên có cấu trúc thành các cụm.

*   `n_samples=300`: Số lượng điểm dữ liệu.

*   `centers=4`: Số lượng cụm.

*   `cluster_std=0.6`: Độ phân tán của các cụm.

*   `random_state=0`: Đảm bảo kết quả nhất quán khi chạy nhiều lần.

3.   **Khởi tạo mô hình K-means:**

*   `n_clusters=4`: Số cụm cần phân loại.

*   `random_state=0`: Đảm bảo tính nhất quán khi chạy lại.

*   `kmeans.fit(X)`: Huấn luyện mô hình với dữ liệu X.

4.   **Dự đoán và lấy tâm cụm:**

*   `kmeans.predict(X)`: Dự đoán nhãn của từng điểm.

*   `kmeans.cluster_centers_`: Lấy tọa độ các tâm cụm sau khi phân cụm.

5.   **Trực quan hóa:**

*   Vẽ các điểm dữ liệu với màu sắc khác nhau tùy theo cụm.

*   Vẽ các tâm cụm bằng biểu tượng X màu đỏ.

# **Code Python cho DBSCAN Clustering:**

```python
# Import các thư viện cần thiết
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

# Tạo dữ liệu mẫu
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.5, random_state=0)

# Trực quan hóa dữ liệu ban đầu
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Dữ liệu ban đầu")
plt.show()

# Khởi tạo mô hình DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)  # eps: bán kính, min_samples: số điểm tối thiểu
labels = dbscan.fit_predict(X)

# Số lượng cụm và các nhãn
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
print(f"Số lượng cụm được tìm thấy: {n_clusters}")

# Trực quan hóa kết quả phân cụm
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.title("DBSCAN Clustering")
plt.show()
```

# **Giải thích từng phần:**
1. **Import thư viện:**
*   `numpy`: Xử lý các phép toán trên mảng.

*   `matplotlib`: Trực quan hóa dữ liệu và kết quả phân cụm.

*   `sklearn.datasets`: Tạo ra các đám mây điểm dữ liệu (blobs).

*   `sklearn.cluster`: Hàm DBSCAN để thực hiện phân cụm.

2. **Tạo dữ liệu mẫu:**
*   Hàm `make_blobs` tạo ra dữ liệu có dạng cụm.

*   `n_samples=300`: Số lượng điểm dữ liệu.

*   `centers=4`: Số cụm trong dữ liệu.

*   `cluster_std=0.5`: Độ phân tán của các cụm.

*   `random_state=0`: Để kết quả không thay đổi mỗi khi chạy lại.

3. **Khởi tạo và huấn luyện mô hình DBSCAN:**
*   `eps=0.3`: Bán kính lân cận để xem xét cụm.

*   `min_samples=5`: Số điểm tối thiểu trong vùng lân cận để tạo thành cụm.

*   `dbscan.fit_predict(X)`: Huấn luyện mô hình và dự đoán nhãn cùng lúc.

*   Nhãn kết quả:

    *   Số cụm được tìm thấy là số nhãn khác nhau trừ đi 1 nếu có nhãn -1 (outlier).

4. **Trực quan hóa kết quả:**
*   Điểm dữ liệu được tô màu theo cụm.

*   Outlier (nếu có) sẽ có nhãn là -1, thường được biểu thị bằng màu khác biệt (màu đen).

5.   **Lựa chọn tham số:**
*   `eps (epsilon)`: Quyết định khoảng cách tối đa giữa các điểm để chúng được coi là hàng xóm.

*   `min_samples`: Số điểm tối thiểu cần có trong vùng lân cận để tạo thành một cụm.

Việc điều chỉnh các tham số này phụ thuộc vào đặc điểm của tập dữ liệu.

# **Code Python cho Spectral Clustering:**

```python
# Import các thư viện cần thiết
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import SpectralClustering

# Tạo dữ liệu dạng hai vòng tròn lồng vào nhau (moons)
X, y = make_moons(n_samples=300, noise=0.1, random_state=0)

# Trực quan hóa dữ liệu ban đầu
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Dữ liệu ban đầu")
plt.show()

# Khởi tạo và huấn luyện mô hình Spectral Clustering
spectral = SpectralClustering(n_clusters=2, affinity='nearest_neighbors', random_state=0)
labels = spectral.fit_predict(X)

# Trực quan hóa kết quả phân cụm
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.title("Spectral Clustering")
plt.show()
```

# **Giải thích từng phần:**
1. **Import thư viện:**
*   `numpy`: Xử lý các phép toán trên mảng.

*   `matplotlib`: Trực quan hóa dữ liệu trước và sau phân cụm.

*   `sklearn.datasets`: Tạo dữ liệu dạng hình trăng khuyết (moons).

*   `sklearn.cluster`: Sử dụng hàm SpectralClustering để phân cụm.

2. **Tạo dữ liệu mẫu:**
*   Hàm `make_moons` tạo ra dữ liệu có dạng hai vòng tròn lồng vào nhau (điển hình cho bài toán khó của K-means).

*   `n_samples=300`: Số lượng điểm dữ liệu.

*   `noise=0.1`: Độ nhiễu của dữ liệu.

*   `random_state=0`: Đảm bảo kết quả nhất quán khi chạy lại.

*   Dữ liệu ban đầu có dạng hai cung trăng khuyết (moons) đan xen nhau, gây khó khăn cho các thuật toán dựa trên khoảng cách như K-means.

3. **Khởi tạo và huấn luyện mô hình Spectral Clustering:**
*   `n_clusters=2`: Số cụm cần tìm (vì dữ liệu có hai vòng tròn).

*   `affinity='nearest_neighbors'`: Tính toán độ tương đồng dựa trên k-nearest neighbors.

*   `random_state=0`: Đảm bảo tính nhất quán khi chạy nhiều lần.

*   `spectral.fit_predict(X)`: Huấn luyện mô hình và dự đoán nhãn cùng lúc.

4. **Trực quan hóa kết quả:**
*   Các điểm thuộc cùng một cụm được tô màu giống nhau.

*   Vì Spectral Clustering sử dụng biểu đồ đồ thị để xác định cụm, nó có khả năng xử lý tốt các dữ liệu không có hình dạng cầu tròn (khác với K-means).

# **Lý do sử dụng Spectral Clustering:**
*   Spectral Clustering thường được sử dụng khi dữ liệu có hình dạng phi tuyến tính hoặc có cấu trúc phức tạp.

*   Khác với K-means, Spectral Clustering không cố gắng tìm cụm hình cầu, mà thay vào đó sử dụng phân tích phổ (eigenvalues) trên biểu đồ đồ thị để phát hiện cụm.

*   Trong ví dụ trên, hai vòng tròn lồng vào nhau (moons) là ví dụ điển hình mà K-means không thể xử lý tốt, nhưng Spectral Clustering phân cụm khá chính xác.

# **Sự khác biệt trước và sau khi phân cụm:**
*   Trước phân cụm: Tất cả điểm có cùng màu, không được gán nhãn.

*   Sau phân cụm: Hai cụm riêng biệt được phân biệt bằng hai màu khác nhau. Spectral Clustering nhận diện đúng hai vòng tròn lồng vào nhau thay vì phân cụm sai lệch như K-means.

# **Code Python cho Mean Shift Clustering:**

```python
# Import các thư viện cần thiết
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import MeanShift, estimate_bandwidth

# Tạo dữ liệu mẫu
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=0)

# Trực quan hóa dữ liệu ban đầu
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Dữ liệu ban đầu")
plt.show()

# Ước lượng bandwidth (băng thông) cho Mean Shift
bandwidth = estimate_bandwidth(X, quantile=0.2, n_samples=300)

# Khởi tạo và huấn luyện mô hình Mean Shift
mean_shift = MeanShift(bandwidth=bandwidth, bin_seeding=True)
labels = mean_shift.fit_predict(X)
cluster_centers = mean_shift.cluster_centers_

# Số lượng cụm được tìm thấy
n_clusters = len(np.unique(labels))
print(f"Số lượng cụm được tìm thấy: {n_clusters}")

# Trực quan hóa kết quả phân cụm
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], s=300, c='red', marker='X', label='Centroids')
plt.title("Mean Shift Clustering")
plt.legend()
plt.show()
```

# **Giải thích từng phần:**
1. **Import thư viện:**
*   `numpy`: Xử lý các phép toán trên mảng.

*   `matplotlib`: Trực quan hóa dữ liệu trước và sau phân cụm.

*   `sklearn.datasets`: Tạo ra các cụm điểm dữ liệu (blobs).

*   `sklearn.cluster`: Sử dụng hàm MeanShift và estimate_bandwidth để phân cụm.

2. **Tạo dữ liệu mẫu:**
*   Hàm `make_blobs` tạo dữ liệu ngẫu nhiên có 4 cụm.

*   `n_samples=300`: Số lượng điểm dữ liệu.

*   `centers=4`: Số lượng cụm.

*   `cluster_std=0.6`: Độ phân tán của các cụm.

*   `random_state=0`: Đảm bảo kết quả nhất quán khi chạy lại.

3. **Ước lượng băng thông:**
*   `estimate_bandwidth`: Tự động xác định độ rộng băng thông (bandwidth) dựa trên phân phối dữ liệu.

*   `quantile=0.2`: Xác định tỷ lệ phần trăm để chọn bandwidth, thường trong khoảng [0.1, 0.3].

*   `n_samples=300`: Số lượng mẫu dùng để ước lượng bandwidth.

4. **Khởi tạo và huấn luyện mô hình Mean Shift:**
*   `MeanShift(bandwidth=bandwidth, bin_seeding=True)`:

   *   `bandwidth`: Độ rộng của kernel (hạt nhân), ảnh hưởng đến cách xác định cụm.

   *   `bin_seeding=True`: Tăng tốc quá trình tìm kiếm cụm.

*   `mean_shift.fit_predict(X)`: Huấn luyện mô hình và dự đoán nhãn cùng lúc.

*   `cluster_centers_`: Lấy tọa độ các tâm cụm sau phân cụm.

5. **Trực quan hóa kết quả:**
*   Các điểm dữ liệu được tô màu dựa trên cụm mà chúng thuộc về.

*   Các tâm cụm (centroids) được đánh dấu bằng dấu X màu đỏ.

*   Hiển thị số lượng cụm được tìm thấy.

# **Tại sao dùng Mean Shift Clustering?**
*   Mean Shift không yêu cầu biết trước số lượng cụm như K-means.

*   Nó tìm cụm dựa trên mật độ của các điểm, giúp phát hiện số cụm tự động.

*   Đặc biệt hữu ích khi các cụm không có dạng hình cầu hoặc có mật độ thay đổi.


# **Code Python cho Hierarchical Clustering:**

```python
# Import các thư viện cần thiết
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Tạo dữ liệu mẫu
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=0)

# Trực quan hóa dữ liệu ban đầu
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Dữ liệu ban đầu")
plt.show()

# Xây dựng mô hình phân cụm phân cấp (Agglomerative Clustering)
hierarchical = AgglomerativeClustering(n_clusters=4, linkage='ward')
labels = hierarchical.fit_predict(X)

# Trực quan hóa kết quả phân cụm
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.title("Hierarchical Clustering")
plt.show()

# Vẽ sơ đồ cây phân cấp (Dendrogram)
linked = linkage(X, method='ward')

plt.figure(figsize=(10, 7))
dendrogram(linked)
plt.title("Dendrogram - Hierarchical Clustering")
plt.show()
```

# **Giải thích từng phần:**
1. **Import thư viện:**
*   `numpy`: Xử lý các phép toán trên mảng.

*   `matplotlib`: Trực quan hóa dữ liệu và kết quả phân cụm.

*   `sklearn.cluster`: Thực hiện phân cụm phân cấp (Agglomerative Clustering).

*   `scipy.cluster.hierarchy`: Tạo dendrogram từ dữ liệu phân cấp.

2. **Tạo dữ liệu mẫu:**
*   Hàm `make_blobs` tạo ra các điểm dữ liệu phân tán có 4 cụm.

*   `n_samples=300`: Số lượng điểm dữ liệu.

*   `centers=4`: Số cụm ban đầu.

*   `cluster_std=0.6`: Độ phân tán của các cụm.

*   `random_state=0`: Đảm bảo kết quả nhất quán khi chạy lại.

*   Biểu đồ ban đầu thể hiện các điểm chưa được phân cụm.

3. **Khởi tạo và huấn luyện mô hình Hierarchical Clustering:**
*   `AgglomerativeClustering`: Thực hiện phân cụm phân cấp theo cách gộp dần (agglomerative).

*   Tham số quan trọng:

   *   `n_clusters=4`: Số cụm muốn tạo ra.

   *   `affinity='euclidean'`: Khoảng cách sử dụng để tính toán sự gần nhau giữa các điểm.

   *   `linkage='ward'`: Phương pháp liên kết Ward để giảm thiểu phương sai trong từng cụm.

*   `fit_predict(X)`: Huấn luyện mô hình và trả về nhãn phân cụm.

4. **Trực quan hóa kết quả phân cụm:**
*   Các điểm được tô màu dựa trên cụm mà chúng thuộc về sau khi phân cụm.

*   Giúp chúng ta thấy được các cụm được phân tách như thế nào.

5. **Vẽ sơ đồ cây phân cấp (Dendrogram)**:
*   `linkage(X, method='ward')`: Tạo ma trận liên kết giữa các điểm.

*   `dendrogram(linked)`: Vẽ biểu đồ dendrogram.

*   Sơ đồ cây cho thấy quá trình gộp các điểm lại với nhau để tạo thành các cụm lớn hơn.

*   Có thể thấy được mối quan hệ phân cấp từ các điểm lẻ cho đến khi hợp nhất thành một cụm lớn.

# **Tại sao sử dụng Hierarchical Clustering?**
*   Không cần biết trước số cụm: Khác với K-means, bạn không cần chỉ định số cụm trước.

*   Cấu trúc phân cấp: Dễ hiểu hơn nhờ sơ đồ dendrogram.

*   Xử lý tốt với dữ liệu có cấu trúc phân cấp hoặc dạng cây.

*   Có thể chọn mức cắt từ dendrogram để xác định số cụm phù hợp.

# **Code Python cho Gaussian Mixture Model (GMM)**

```python
# Import các thư viện cần thiết
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.mixture import GaussianMixture

# Tạo dữ liệu mẫu
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=0)

# Trực quan hóa dữ liệu ban đầu
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Dữ liệu ban đầu")
plt.show()

# Khởi tạo và huấn luyện mô hình Gaussian Mixture Model (GMM)
gmm = GaussianMixture(n_components=4, covariance_type='full', random_state=0)
gmm.fit(X)

# Dự đoán nhãn cho từng điểm dữ liệu
labels = gmm.predict(X)

# Trực quan hóa kết quả phân cụm
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.title("Gaussian Mixture Clustering")
plt.show()

# In ra xác suất mỗi điểm thuộc từng cụm
probs = gmm.predict_proba(X)
print("Xác suất mỗi điểm thuộc từng cụm:")
print(probs[:5])  # Hiển thị 5 dòng đầu tiên
```

*   `sklearn.mixture`: Sử dụng GaussianMixture để thực hiện phân cụm theo mô hình hỗn hợp Gaussian.

**2. Tạo dữ liệu mẫu:**

*   `random_state=0`: Đảm bảo kết quả nhất quán khi chạy lại.

*   Biểu đồ ban đầu hiển thị các điểm chưa phân cụm.

**3. Khởi tạo và huấn luyện mô hình GMM:**
*   `n_components=4`: Số lượng Gaussian (cụm) muốn tìm.

*   `gmm.fit(X)`: Huấn luyện mô hình trên dữ liệu.

**4. Dự đoán nhãn của từng điểm:**
*   `gmm.predict(X)`: Dự đoán cụm mà mỗi điểm thuộc về.

*   Các điểm thuộc cùng cụm sẽ được tô màu giống nhau trong biểu đồ.

**5. Trực quan hóa kết quả phân cụm:**
*   Các điểm dữ liệu được tô màu dựa trên cụm mà chúng thuộc về.

*   GMM phân cụm theo phân phối Gaussian, do đó các cụm có dạng hình ellipse thay vì hình cầu như K-means.

**6. In xác suất mỗi điểm thuộc từng cụm:**
*   `gmm.predict_proba(X)`: Tính xác suất mỗi điểm thuộc về từng cụm.

*   In ra 5 dòng đầu để xem tỷ lệ phân phối xác suất.

**Tại sao dùng Gaussian Mixture Model (GMM)?**
*   **Phân phối Gaussian**: Các cụm có dạng ellipse thay vì hình cầu.

*   **Xác suất**: GMM không chỉ đưa ra nhãn mà còn cho biết xác suất thuộc về từng cụm.

*   **Linh hoạt**: Xử lý các cụm có hình dạng khác nhau.

*   **Xử lý dữ liệu phức tạp**: GMM có thể xử lý các cụm chồng lấn (overlapping).
